```python
#question 2
*args and **kwargs are used to take any number of inputs in a function. *args when used returns a tuple containing the input variables , while **kwargs is a special keyword which when used gives dictionary of the input variables.

eg- 
def sum_numbers(*args):
    return sum(args)

print(sum_numbers(10,20,30,40))


eg-
def intro(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")

intro(name = "Alice",role="Developer",location="india")    
```

```python
#question 4
A generator function in Python is a special type of function that returns an iterator object instead of a single value. It uses yield keyword which basically tells to store that value locally and then continue on to the next iteration.
eg-
def fib(n):
    a,b = 0,1;
    for i in range(n):
        yield a
        a,b = b , a+b
```

```python
fib(10)
```

```python
#question 5
def prime_generator():
    for num in range(2, 1000):
        is_prime = True
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                is_prime = False
                break

        if is_prime:
            yield num
primes = prime_generator()
for _ in range(20):
    print(next(primes))
```
