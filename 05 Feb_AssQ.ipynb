```python
#q5
Inheritance is a mechanism in OOP where a new class (derived/child class) acquires the properties and behaviors (attributes and methods) of an existing class (base/parent class).

1)single inheritance
eg- 
class Animal:       # Parent
    def eat(self): print("Eating...")

class Dog(Animal):  # Child
    def bark(self): print("Barking...")

2)multiple inheritance
eg - class Father:
    def skill_a(self): print("Driving")

class Mother:
    def skill_b(self): print("Cooking")

class Child(Father, Mother): # Inherits from both
    pass

3)multilevel inheritance
eg-
class Vehicle:       # Grandparent
    def start(self): print("Engine started")

class Car(Vehicle):  # Parent
    def drive(self): print("Car driving")

class ElectricCar(Car): # Child
    def charge(self): print("Battery charging")

4)hierarchical inheritance
eg-
class Shape: 
    def area(self): print("Calculating area...")

class Circle(Shape): # Child A
    pass

class Square(Shape): # Child B
    pass

```
