```python
#q3
multiple inheritance is type of interitance when a child inherits its property from more than one parent.
```

```python
class parent1():
    def type1(self):
        print("hi")

class parent2():
    def type2(self):
        print("hello")

class child(parent1,parent2):
    pass
    
```

```python
c1 = child()
c1.type1()
c1.type2()
```

```python
#q4
getter and setter are methods to access and change the private/protected variables of a class variable
```

```python
class bank_account():
    def __init__(self,balance):
        self.__balance = balance

    @property
    def balance(self):
        print("Fetching balance...")
        return self.__balance

    @balance.setter
    def balance(self, amount):
        if amount < 0:
            print("Error: Balance cannot be negative!")
        else:
            print("Updating balance...")
            self._balance = amount
```

```python
account = bank_account(1000)
print(account.balance)

account.balance = 1500
print(account.balance)

account.balance = -500
```

```python
#q5
Method overriding happens when a child class provides a specific implementation of a method that is already defined in its parent class. The child's method "overrides" the parent's version when called on a child instance.
```

```python
class animal():
    def speak(self):
        print("aaaaaaoooooo")

class cat(animal):
    def speak(self):
        print("meow")
```

```python
c = cat()
c.speak()
```
