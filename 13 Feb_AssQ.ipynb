```python
#q3---> arithmaticError class is a built in class used to handle errors that occur during mathematical calculations. eg include
#ZeroDivisionError
try:
    x = 10 / 0
except ZeroDivisionError:
    print("Error: You cannot divide by zero!")
#OverflowError
try:
    result = 10.0 ** 10000
except OverflowError:
    print("Error: The calculation exceeded the representable limit!")
```

```python
#q4---> the LookupError class in python serves as the base class for exceptions raised when a key or index used on a mapping or sequence is invalid.
#keyerror
user_profile = {"name": "Alice", "age": 30}

try:
    email = user_profile["email"]
except KeyError as e:
    print(f"Error: The key {e} was not found in the dictionary.")

#indexerror
items = ["apple", "banana", "cherry"]

try:
    fourth_item = items[5]
except IndexError as e:
    print(f"Error: {e}. The index is out of range.")

#LookupError
def get_data_safely(collection, key_or_index):
    try:
        return collection[key_or_index]
    except LookupError as e:
        return f"Lookup failed: {type(e).__name__} occurred."

print(get_data_safely({"a": 1}, "b"))  

print(get_data_safely([10, 20], 5))    


```
