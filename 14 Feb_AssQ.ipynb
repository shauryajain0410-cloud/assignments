```python
#q1---> Suppose we write a function , suppose 'test' function and then we ran it for various inputs like test(1),test(2) and so on , so for very large number of inputs , it would be a tedious task , what we do is we make a sort of thread in which we have to write only once and then the function would be ran for all inputs. The module used to do so is called threading
```

```python
#q2---> threading module is used to handle threads in python.
#1) activeCount(): returns the total no of thread objects that are currently there
#2) currentThread(): returns the specific thread object representing the caller's current thread of control.
#3) enumerate(): Returns a list of all Thread objects currently alive.
```

```python
#q3 ---> start(): Starts the thread's activity in a separate, independent thread of control.    run(): Represents the actual entry point and target activity that executes inside the thread.  join():Blocks the calling thread (usually the main program) until the target thread terminates.  isAlive(): Returns a boolean value (True or False) indicating whether the thread is currently executing. 
```

```python
#q4
import threading
def func_sq(x):
    print (x*x)

def func_cub(x):
    print (x*x*x)
```

```python
thread1 = [threading.Thread(target = func_sq, args = (i,)) for i in [2,4,6,8,10]]
thread2 = [threading.Thread(target = func_cub, args = (i,)) for i in [1,3,5,7,9]]
```

```python
for t in thread1:
    t.start()
```

```python
for t in thread2:
    t.start()
```

```python
#q5----> advantages - multiple thtreads can keep the CPU busy and reduce idle time, one thread can execute execution while other waiting for I/O operations
#disadvantages - debugging multithreads are difficult, deadlocks , Frequent switching between threads can reduce performance.
```

```python
#q6 ---> A deadlock occurs when two or more threads are waiting indefinitely for resources held by each other, causing the program to stop progressing.
#A race condition occurs when multiple threads access and modify shared data simultaneously, and the final result depends on the order of execution.
```
