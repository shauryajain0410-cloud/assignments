```python
#q1 ---> flask is a lightweight open source python web framework designed to make building web applications and API's.Its advantages are that it is lightweight , minimalistic , easy to learn , built in local development server
```

```python
#q2 --->
from flask import Flask
app = Flask(__name__)

@app.route('/1')
def display():
    return 'Hello World!!'

if __name__ == "__main__":
    app.run(host="0.0.0.0")
```

```python
#q3 ---> app routing is process of mapping a specific url to the python function. We use it so that the users see the specific content , data or webpage they requested. it is done by the app.route decorator used above also.
```

```python
#q4
from flask import Flask
app = Flask(__name__)

@app.route('/welcome')
def yo():
    return 'Welcome to ABC Corporation'

@app.route('/')
def yo1():
    return """Company Name: ABC Corporation
            Location: India
            Contact Detail: 999-999-9999"""

if __name__ == "__main__":
    app.run(host="0.0.0.0")
```
