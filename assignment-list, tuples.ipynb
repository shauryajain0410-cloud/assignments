```python
#q2
the two in-built functions are:
a) count()
t = (1,3,8,9,8,8,5,6,8,5)
print(t.count(5))
b) index()
fruits = ("apple", "banana", "cherry")
print(fruits.index("banana"))

Tuples only have two methods because of their immutability
```

```python
#q4
union() returns a new set containing all elemnnts from both sets without changing the original sets , while update() returns a set with elements of one added to the other , and here the set is now modified.
set1 = {1, 2, 3}
set2 = {3, 4, 5}
new_set = set1.union(set2)
print(new_set)
print(set1) 

set1.update(set2)
print(set1) 
```
